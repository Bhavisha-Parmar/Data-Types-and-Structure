## **2.  Explain the difference between mutable and immutable data types with examples?**

**In Python, Mutable data types** can be modified after creation (like lists
and dictionaries), while **immutable data types** cannot (like strings and tuples).

**Mutable Data Types:**
  These data types allow you to change their values after they are created, without creating a new object.

* **Examples:**

**Lists:** my_list = [1, 2, 3] can be modified by adding, removing, or changing elements.

**Dictionaries:** my_dict = {'a': 1, 'b': 2} can be modified by adding, removing, or changing key-value pairs.

**Sets:** my_set = {1, 2, 3} can be modified by adding or removing elements.

* **Example:**

  my_list = [1, 2, 3]

  my_list[0] = 4  # Modifying the first element

  print(my_list)  # Output: [4, 2, 3]


**Immutable Data Types:**
These data types cannot be changed after they are created, meaning any operation that appears to modify them actually creates a new object with the modified value.

* **Examples:**

**Strings:** my_string = "hello" cannot be changed directly; any modification creates a new string object.

**Tuples:** my_tuple = (1, 2, 3) cannot be changed after creation.

**Numbers (int, float, etc.):** my_number = 5 cannot be changed; any operation creates a new number object.

* **Example:**

my_string = "hello"

my_string[0] = 'w'  //This will raise an error because strings are immutable

new_string = my_string + " world" # Creating a new string

print(new_string) # Output: hello world

##**8.  What is a hash table, and how does it relate to dictionaries in Python?**

**What is a Hash Table?**

* A hash table is a data structure that stores key-value pairs.

* It uses a hash function to determine the storage location (index) of a key within an array (or "buckets").

* This allows for efficient (average O(1) time complexity) retrieval, insertion, and deletion of data.

**How it relates to Python Dictionaries:**

* Python dictionaries are implemented using hash tables.

* When you create a dictionary, Python uses a hash function to map the keys to indices within an internal array (or "buckets").

* This allows you to access values in a dictionary very quickly, based on their keys.

## **11. What advantages do dictionaries offer over lists for certain tasks?**

Dictionaries excel over lists in Python for tasks requiring fast lookups and storing data as key-value pairs, offering efficient retrieval and organization of information.

* **Fast Lookups:**

Dictionaries use a hash table (or hash map) internally, allowing for near-instantaneous retrieval of values based on their keys, while lists require sequential searching.

* **Key-Value Association:**

Dictionaries store data as key-value pairs, making it easy to associate data logically and retrieve specific information using a unique key.

* **Flexibility:**

Dictionaries can store any data type as values, including other dictionaries or lists, providing flexibility in structuring complex data.

* **Memory Efficiency:**

Dictionaries only store the key-value pairs, unlike lists which store all elements sequentially, potentially saving memory when dealing with large datasets.

* **Built-in Methods:**

Python provides a rich set of built-in methods for manipulating dictionaries, such as get(), keys(), values(), and items(), simplifying common operations.

* **Mapping Relationships:**

Dictionaries are well-suited for mapping relationships between items, such as storing user data (name, age, email) where the names act as keys and the other data as values.

* **Readability:**

Dictionaries make code more readable when dealing with key-value pairs, as the association between data is explicit.

* **Unordered:**

Dictionaries are unordered, meaning the order of key-value pairs is not preserved, which can be a disadvantage in certain situations where order is important.

## **14.  How does the in keyword work differently for lists and dictionaries?**

* The in operator tells you whether something appears as a key in the dictionary. It returns two values True or False.

* The in operator uses different algorithms for lists and dictionaries. For lists, it uses a search algorithm. As the list gets longer, the search time gets longer in direct proportion.

* For dictionaries, Python uses an algorithm called a hashtable that has a remarkable property: the in operator takes about the same amount of time no matter how many items there are in a dictionary.

## **16.  What is a nested dictionary, and give an example of its use case?**

* A nested dictionary is a dictionary where the values are themselves dictionaries, allowing for hierarchical data organization.

* A use case is representing an organizational chart, where departments (keys) map to dictionaries containing employee IDs and names (values).

**Example:**

    company_org_chart = {
      "Marketing": {
        "ID234": "Jane Smith",
        "ID567": "John Doe"
      },
      "Sales": {
        "ID123": "Bob Johnson",
        "ID122": "David Lee"
      },
      "Engineering": {
        "ID303": "Radhika Potlapally",
        "ID321": "Maryam Samimi"
      }
    }

**In this example:**

* company_org_chart is the main dictionary.

* The keys are department names ("Marketing", "Sales", "Engineering").

* The values are dictionaries containing employee information (ID and name).

* To access an employee's name, you would use double brackets:

  company_org_chart["Sales"]["ID122"] would return "David Lee".



## **17. Describe the time complexity of accessing elements in a dictionary.**

The time complexity of retrieving a value by its key in a dictionary is O(1). This is because dictionaries use a hash table internally which allows constant time lookup.



## **18.  In what situations are lists preferred over dictionaries?**

* A list can store any data type. In contrast, keys in the dictionary can be of any immutable data type, and values can be of any data type. Lists perform faster for ordered operations like sorting, while dictionaries perform faster for lookup operations.

* For quick data look-ups, configurations, or caches, favor dictionaries. For ordered collections and sequence operations, such as maintaining a stack or queue, lists are more suitable.

## **19. Why are dictionaries considered unordered, and how does that affect data retrieval?**

Dictionaries are considered unordered because they don't inherently maintain the order in which key-value pairs are inserted, unlike lists or tuples. This lack of inherent order means you can't rely on the order of keys when iterating or retrieving data.

## **20. Explain the difference between a list and a dictionary in terms of data retrieval.**

In data retrieval, lists access elements by their numerical index (starting from 0), while dictionaries access elements by their unique keys, offering potentially faster lookups for large datasets.
